```python
plt.figure(figsize=(6,4))
plt.hist(df["population"], bins=15)
plt.title("Distribution of Population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualization 2 - Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="gdp",
                y="population",
                hue="continent",
                data=df)

plt.title("GDP vs Population by Continent")
plt.xlabel("GDP")
plt.ylabel("Population")
plt.show()
```

```python
 #Visualization 3 - Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="continent",
            y="gdp",
            data=df)

plt.title("GDP by Continent")
plt.xlabel("Continent")
plt.ylabel("GDP")
plt.show()
```

```python
# Visualization 4 - Bar Chart
plt.figure(figsize=(6,4))
df["continent"].value_counts().plot(kind="bar")

plt.title("Number of Countries in Each Continent")
plt.xlabel("Continent")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
