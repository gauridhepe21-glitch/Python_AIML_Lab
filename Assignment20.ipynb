```python
import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("ford_car_dataset.csv")
```

```python
X = df.drop("price", axis=1)
Y = df["price"]
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
categorical_columns = X.select_dtypes(include=['object']).columns

print("\nCategorical Columns:")
print(categorical_columns)

X = pd.get_dummies(X, columns=categorical_columns).astype(int)

print("\nFirst 5 Rows After Encoding:")
print(X.head())
```

```python
scaler = StandardScaler()

numerical_columns = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

for col in numerical_columns:
    if col in X.columns:
        X[col] = scaler.fit_transform(X[[col]])

print("\nFirst 5 Rows After Scaling:")
print(X.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.33,
    random_state=42
)

print("\nTraining Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape:", y_test.shape)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)

print("\nIntercept:")
print(model.intercept_)

print("\nCoefficients:")
print(model.coef_)
```

```python
y_pred = model.predict(X_test)

print("\nFirst 10 Predicted Values:")
print(y_pred[:10])

print("\nFirst 10 Actual Values:")
print(y_test.values[:10])
```

```python
r2 = r2_score(y_test, y_pred)

print("\nR² Score:", r2)

if r2 > 0.9:
    print("Excellent model performance.")
elif r2 > 0.75:
    print("Good model performance.")
elif r2 > 0.5:
    print("Average model performance.")
else:
    print("Poor model performance.")
```

```python
joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(list(X.columns), "columns.pkl")

print("\nModel and preprocessing files saved successfully!")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
```

```python
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Root Mean Squared Error (RMSE):", rmse)

print("\nModel Evaluation Summary")
print("R² Score :", r2)
print("MAE      :", mae)
print("MSE      :", mse)
print("RMSE     :", rmse)
```
